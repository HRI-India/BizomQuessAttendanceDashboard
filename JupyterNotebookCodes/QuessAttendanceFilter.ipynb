```python
import pandas as pd
```

```python
# Load both sheets to inspect their structure
summary_df = pd.read_excel("/content/Quess_June.xlsx", sheet_name='Summary')
Attendance_df = pd.read_excel("/content/Quess_June.xlsx", sheet_name='Attendance')

# Display the first few rows of each to understand the structure
summary_df.head(), Attendance_df.head()
```

```python
# First, check all column names to identify the date columns
all_columns = Attendance_df.columns

# Attempt to parse column names as dates to filter those between 21 April and 20 May 2025
from datetime import datetime

date_columns = []
start_date = datetime(2025, 5, 21)
end_date = datetime(2025, 6, 20)

# Identify date columns by trying to parse column names as dates
for col in all_columns:
    try:
        col_date = pd.to_datetime(col, errors='raise')
        if start_date <= col_date <= end_date:
            date_columns.append(col)
    except:
        continue

date_columns
```

```python
# Convert datetime objects to strings matching original column labels
date_str_columns = [d.strftime('%Y-%m-%d') for d in date_columns]

# First, rename matching date columns in the DataFrame to ensure melt works
rename_map = dict(zip(date_columns, date_str_columns))
Attendance_df_renamed = Attendance_df.rename(columns=rename_map)

# Define ID variables (non-date columns to retain in the reshaped data)
id_vars = [col for col in Attendance_df_renamed.columns if col not in date_str_columns]

# Melt the DataFrame from wide to long format
reshaped_df = pd.melt(
    Attendance_df_renamed,
    id_vars=id_vars,
    value_vars=date_str_columns,
    var_name='Date',
    value_name='Attendance Status'
)

# Save the reshaped data to a new Excel file
output_path = "Reshaped_Quess_Attendance_May21_June20.xlsx"
reshaped_df.to_excel(output_path, index=False)

output_path
```
